```python
# Dot Product of two vectors

a = np.array([4,5])
b = np.array([6,4])

a_b_dot_product = np.dot(a,b)

print(a_b_dot_product)
```

```python
# Cross Product of two vectors

c = np.array([2,3,0])
d = np.array([4,4,0])

c_d_cross_product = np.cross(c,d)

print(c_d_cross_product)
```

```python
# Projection of a vector
'''Formula: proj_of_a_vector = ((a dot v)/|v**2|))*v'''

a = np.array([23,45,62])
v = np.array([45,82,67])

a_dot_v = np.dot(a,v)

magnitude_of_v = np.sqrt(sum(v**2))

proj_of_a_vector = (a_dot_v / magnitude_of_v**2) * v

print(proj_of_a_vector)
```
